```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots

from ipywidgets import interact, widgets
from datetime import datetime
```

```python
# Usamos el dataset generado en la extraccion de datos
# definimos el archivo .csv a usar

df_tanks = pd.read_csv("API_20231.csv", index_col='date')
df_tanks.head()
```

```python
# Definimos la funcion para graficar usando plotly

def plot_columns(columns):
    fig = make_subplots(rows=1, cols=1)
    for column in columns:
        trace = go.Scatter(x=df_tanks.index, y=df_tanks[column], mode='lines', name=column)
        fig.add_trace(trace)

    fig.update_xaxes(title_text='Fecha')
    fig.update_yaxes(title_text='Gravedad API') # Cambiar nombre del eje vertical >> %BSW

    # Ajustar los valores del eje y
    fig.update_yaxes(range=[12, 20])

    fig.update_layout(width=1000, height=800)
    
    fig.update_layout(title='Gráfico de Datos', showlegend=True)
    fig.update_xaxes(rangeslider_visible=True)

    #fig.add_vrect(x0='2023-11-20', x1='2023-11-20')
    #fig.add_vline(x='2023-11-20', line_width=3, line_dash="dash", line_color="green")
    
    # Linea vertical

    #fig.add_vrect(x0="2023-11-20", x1="2023-11-20", 
              #annotation_text="Despresurizacion, posible taponamiento URV", annotation_position="top left",
              #fillcolor="orange", opacity=1, line_width=3, line_dash="dash", line_color="orange")

    fig.show()
```

```python
# Obtener la lista de columnas disponibles para graficar
columnas_disponibles = df_tanks.columns

# Crear un widget interactivo para seleccionar múltiples columnas a graficar
interact(plot_columns, columns=widgets.SelectMultiple(options=columnas_disponibles, description='Columnas:'))
```
